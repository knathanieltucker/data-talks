## Overall Question: Do critics rate action movies lower than users?

Let's break this down into steps to make it more manangeable and allow us to rigorously investigate this question.

1. Examine the data

  - Define columns to use

  - Create necessary variables

  - Check for data quality issues

3. Clean and fix anything that we need to

4. Exploratory plots

  - Can we notice any differences visually between critic and user ratings?

4. Start simple - basic comparison of critic vs. user ratings

  - t test of averages for overall, action, and non-action movies
  
  - Is the difference between average critic and user ratings significant for each movie type?
  
5. More rigorous and direct - compare difference in critic and user ratings for action/non-action movies

  - Linear regression of rating difference vs. action/non-action

  - Is the difference between critic and user ratings significantly different for action movies relative to non action movies?
  
7. Do some robustness checks

  - Is time a confounding factor?

  - Visualize averages over time for critic and user ratings and their difference

  - Does the difference between critic and user ratings change over time for overall, action, and non-action movies?
  
  - Regression of rating difference vs. action/non-action with time trend

```python
# put user rating on same scale
cleaned['RTUserRatingMod'] = cleaned.RTUserRating * 2
```

```python
# compare options for identifying action movies
(cleaned.Genre.str.contains('Action')).sum()
(cleaned.Genre == 'Action').sum()
```

```python
# create isAction variable
cleaned['isAction'] = cleaned.Genre.str.contains('Action')
```

```python
# examine data again
cleaned.head()

cleaned.describe()

cleaned.info()
```

```python
#look at boxplots and violin plots to compare critic and using ratings across movie type
#visualize to get a sense if a difference exists
sns.boxplot('isAction','RTCriticRating', data=cleaned[['isAction', 'RTCriticRating']])
```

```python
sns.boxplot('isAction','RTUserRatingMod', data=cleaned[['isAction', 'RTUserRatingMod']])
```

```python
sns.violinplot('isAction','RTCriticRating', data=cleaned[['isAction', 'RTCriticRating']])
```

```python
sns.violinplot('isAction','RTUserRatingMod', data=cleaned[['isAction', 'RTUserRatingMod']])
```

```python
# compare distribution and correlation of critic and user ratings to see if very different
sns.pairplot(data=cleaned[['RTUserRatingMod', 'RTCriticRating', 'isAction']], 
             kind='reg', hue='isAction', vars=['RTUserRatingMod', 'RTCriticRating'])
```

```python
# simple t test to see if critic and user ratings are significantly different overall
from scipy.stats import ttest_ind

res = ttest_ind(cleaned['RTUserRatingMod'], cleaned['RTCriticRating'], equal_var=False)

res.pvalue
```

```python
# simple t test to see if critic and user ratings are significantly different for action movies

res = ttest_ind(cleaned[cleaned['isAction']]['RTUserRatingMod'], 
                cleaned[cleaned['isAction']]['RTCriticRating'], equal_var=False)

res.pvalue
```

```python
# simple t test to see if critic and user ratings are significantly different for non action movies

res = ttest_ind(cleaned[cleaned['isAction']==False]['RTUserRatingMod'], 
                cleaned[cleaned['isAction']==False]['RTCriticRating'], equal_var=False)

res.pvalue
```

```python
from statsmodels import regression
```

```python
# regression to see if difference for action movies is significantly different than difference for non action movies
cleaned['diff'] = cleaned.RTUserRatingMod - cleaned.RTCriticRating

lm = regression.linear_model.OLS.from_formula(formula='diff ~ 1 + isAction', data=cleaned)
```

```python
res = lm.fit()
```

```python
print res.summary()
```

```python
# group by year for time series plots
gb = cleaned.groupby(by='Year')
```

```python
# examine average user rating by year
sns.tsplot(gb.RTUserRatingMod.mean(), time=gb.indices.keys())
```

```python
# examine average critic rating by year
sns.tsplot(gb.RTCriticRating.mean(), time=gb.indices.keys())
```

```python
# examine average difference in user and critic rating by year
sns.tsplot(gb.RTUserRatingMod.mean() - gb.RTCriticRating.mean(), time=gb.indices.keys())
```

```python
# group by year for time series plots
gba = cleaned[cleaned['isAction']].groupby(by='Year')

sns.tsplot(gba.RTUserRatingMod.mean() - gba.RTCriticRating.mean(), time=gba.indices.keys(), ci=70)
```

```python
# limit to after 1997 because very few movies before then
new_movies = cleaned[cleaned['Year'] > 1997]

new_movies.describe()
```

```python
# difference between user and critic ratings by year for recent action movies
# note the trend where the difference seems to increase over time, so the isAction effect we saw above could be due to this
action_new = new_movies[new_movies['isAction']]
action_new['diff'] = action_new.RTUserRatingMod - action_new.RTCriticRating

sns.boxplot(x='Year', y='diff', data=action_new)
```

```python
# regression to see if difference for action movies is significantly different than non action movies
# add in time trend to see if changes anything as a robustness check
# want to see if coefficient on isAction changes
lm = regression.linear_model.OLS.from_formula(formula='diff ~ 1 + isAction + Year', data=cleaned)
```

```python
res = lm.fit()
```

```python
print res.summary()
```
